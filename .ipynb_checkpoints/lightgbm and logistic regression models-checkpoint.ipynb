```python
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
from sklearn import model_selection
import lightgbm

from sklearn.model_selection import RepeatedStratifiedKFold 
from sklearn.model_selection import GridSearchCV 
from sklearn.linear_model import LogisticRegression
import time
#import optuna
```

```python
current_path = os.path.abspath(os.path.dirname("CDA Project"))
data=pd.read_csv(current_path+r'/source/main_data.csv')
var_columns=[c for c in data.columns if c not in ['TARGET']]
X=data.loc[:,var_columns]
y=data.loc[:,'TARGET']
```

```python
#split data up
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
#do some feature selection, we dont need ALL the columns
#use randomforest
random_forest = RandomForestClassifier(n_estimators = 100, random_state = 2340).fit(X_train, y_train)
feature_select = SelectFromModel(random_forest).fit(X_train, y_train)
```

```python
X_train_transform=pd.DataFrame(feature_select.transform(X_train))
```

```python
X_train
```

```python
#check whether binary classification balance for assigning is_unbalance parameter
amount_1=np.count_nonzero(y_train.to_numpy()==1)
amount_0=np.count_nonzero(y_train.to_numpy()==0)
print(amount_1)
print(amount_0)
print("binary classification unbalanced")
```

```python
X_train = X_train.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
X_test = X_test.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
train_data=lightgbm.Dataset(X_train,label=y_train)
test_data=lightgbm.Dataset(X_test,label=y_test)
```

```python
parameters={
    'objective':'binary',
    'metric':'auc',
    'is_unbalance':'true',
    'boosting':'goss',
    'num_leaves':32,
    'max_depth':5,
    'feature_fraction':0.8,
    'bagging_freq':5,
    'learning_rate':0.01
    
}
```

## LightGBM

```python
#build inital lightgbm model
model=lightgbm.train(
    parameters,
    train_data,
    valid_sets=test_data,
    num_boost_round=300,
    early_stopping_rounds=30
)
```

```python
#predict the probability of the target
y_pred_prob=model.predict(X_test)
```

```python
clf=lightgbm.LGBMClassifier(**parameters)
clf.fit(X_train,y_train)
```

```python
#predict the value of the target
y_pred=clf.predict(X_test)
```

```python
print("Accuracy:%f"%(1-(y_test.to_numpy()!=y_pred).sum()/y_test.to_numpy().shape[0]))
```

```python
#finding the relationship between number of leaves and accuracy
num_leaves=[2,4,8,16,32]
train_score_num_leaves=[]
test_score_num_leaves=[]
for i in num_leaves:
    params={
        'objective':'binary',
        'metric':'auc',
        'is_unbalance':'true',
        'boosting':'goss',
        'num_leaves':i,
        'max_depth':5,
        'feature_fraction':0.8,
        'bagging_freq':5,
        'learning_rate':0.01
    }
    
    clf=lightgbm.LGBMClassifier(**params)
    clf.fit(X_train,y_train)
    
    train_score_num_leaves.append(clf.score(X_train,y_train))
    test_score_num_leaves.append(clf.score(X_test,y_test))

plt.plot(num_leaves,test_score_num_leaves,label="testing accuracy")
plt.plot(num_leaves,train_score_num_leaves, label="training accuracy")
plt.legend(labels=['testing accuracy','training accuracy'], loc='upper right')
plt.xlabel("number of leaves")
plt.ylabel("accuracy")
plt.title("num_leaves=2,4,8,16,32")
```

```python
#overfitting experiment
num_leaves=[32,64,128,256,512]
train_score_num_leaves=[]
test_score_num_leaves=[]
for i in num_leaves:
    params={
        'objective':'binary',
        'metric':'auc',
        'is_unbalance':'true',
        'boosting':'goss',
        'num_leaves':i,
        'max_depth':128,
        'feature_fraction':0.8,
        'bagging_freq':5,
        'learning_rate':0.01
    }
    
    clf=lightgbm.LGBMClassifier(**params)
    clf.fit(X_train,y_train)
    
    train_score_num_leaves.append(clf.score(X_train,y_train))
    test_score_num_leaves.append(clf.score(X_test,y_test))

plt.plot(num_leaves,test_score_num_leaves,label="testing accuracy")
plt.plot(num_leaves,train_score_num_leaves, label="training accuracy")
plt.legend(labels=['testing accuracy','training accuracy'], loc='upper right')
plt.xlabel("number of leaves")
plt.ylabel("accuracy")
plt.title("num_leaves=32,64,128,256,512/ max_depth=128")
```

```python
# find the relationship between learning rate and accuracy
learning_rate=[0.01,0.02,0.05,0.1,0.5]
train_score_learning_rate=[]
test_score_learning_rate=[]
for i in learning_rate:
    params={
        'objective':'binary',
        'metric':'auc',
        'is_unbalance':'true',
        'boosting':'goss',
        'num_leaves':32,
        'max_depth':5,
        'feature_fraction':0.8,
        'bagging_freq':5,
        'learning_rate':i
    }
    
    clf=lightgbm.LGBMClassifier(**params)
    clf.fit(X_train,y_train)
    
    train_score_learning_rate.append(clf.score(X_train,y_train))
    test_score_learning_rate.append(clf.score(X_test,y_test))

plt.plot(learning_rate,test_score_learning_rate, label="testing accuracy")
plt.plot(learning_rate,train_score_learning_rate, label="training accuracy")
plt.legend(labels=['testing accuracy','training accuracy'], loc='upper right')
plt.xlabel("learning rate")
plt.ylabel("accuracy")
plt.title("learning_rate=0.01,0.02,0.05,0.1,0.5")
```

```python
# find the relationship between maximum depth and accuracy
max_depth=[1,3,5,7,9]
train_score_max_depth=[]
test_score_max_depth=[]
for i in max_depth:
    params={
        'objective':'binary',
        'metric':'auc',
        'is_unbalance':'true',
        'boosting':'goss',
        'num_leaves':32,
        'max_depth':i,
        'feature_fraction':0.8,
        'bagging_freq':5,
        'learning_rate':0.01
    }
    print(params)
    
    clf=lightgbm.LGBMClassifier(**params)
    clf.fit(X_train,y_train)
    
    train_score_max_depth.append(clf.score(X_train,y_train))
    test_score_max_depth.append(clf.score(X_test,y_test))

plt.plot(max_depth,test_score_max_depth,label="testing accuracy")
plt.plot(max_depth,train_score_max_depth, label="training accuracy")
plt.legend(labels=['testing accuracy','training accuracy'], loc='upper right')
plt.xlabel("max_depth")
plt.ylabel("accuracy")
plt.title("max_depth=1,3,5,7,9")
```

```python
param_grid={
    'max_depth':[1,3,5],
    'num_leaves':[2,8,32],
    'learning_rate':[0.01,0.05,0.1],
    'boosting': ['goss']
}
```

```python
#using gridsearch to tune the parameters
grid_model=model_selection.GridSearchCV(
    estimator=clf,
    param_grid=param_grid,
    scoring='accuracy',
    verbose=10,
    cv=3,
    n_jobs=-1,
)
```

```python
grid_model.fit(X_train,y_train)
```

```python
print("accuracy:%f"%(grid_model.best_score_))
print(grid_model.best_estimator_.get_params())
```
